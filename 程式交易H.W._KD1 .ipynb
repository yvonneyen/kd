```python
import openpyxl
import pandas as pd
import os
import datetime
#from IPython.display import display
```

```python
excelname= "kd"
sheetname = "kd"

wb = openpyxl.load_workbook(str(excelname)+".xlsx")
ws = wb[str(sheetname)]
```

```python
dataset = []
i = 1
while(ws.cell(row= i, column= 1).value):
    row = []
    j = 1
    while(ws.cell(row= i, column= j).value):
        if j <6:
            row.append(ws.cell(row= i, column= j).value)
        j = j + 1
    dataset.append(row)
    i=i+1
```

```python
pd.DataFrame(dataset)
```

```python
dataset[0].append("9日內最高價格")
dataset[0].append("9日內最低價格")
dataset[0].append("RSV(9)")
dataset[0].append("K(9)")
dataset[0].append("D(9)")
dataset[0].append("buy_day")
dataset[0].append("cum_buy_day")
dataset[0].append("sell_day")
#在列表添加新ㄉ項目
```

```python
i=9
while i< len(dataset):
    highest=[]
    lowest=[]
    j=0
    while j <9:
        highest.append(dataset[i-j][2])
        lowest.append(dataset[i-j][3])
        j=j+1
    dataset[i].append(max(highest))
    dataset[i].append(min(lowest))
    i=i+1
#最近九日最低價\最近九日最高價
```

```python
i=9
while i < len(dataset):
    dataset[i].append((dataset[i][4]-dataset[i][6])/(dataset[i][5]-dataset[i][6])*100)
    i=i+1
#RSV(t) = (今日收盤價 - 最近九日最低價)/ (最近九日最高價 - 最近九日最低價)*100
```

```python
dataset[9].append(50)
dataset[9].append(50)
#K值 初始值 = 50, at day 9
#D值 初始值 = 50, at day 9
```

```python
i=10
while i < len(dataset):
    dataset[i].append(dataset[i-1][8]*(2/3)+dataset[i][7]*(1/3))
    i+=1
#K(t)= K(t-1)*2/3 + RSV(t)*1/3, t = 10, 11, ..., T
```

```python
i=10
while i < len(dataset):
    dataset[i].append(dataset[i-1][9]*(2/3)+dataset[i][8]*(1/3))
    i+=1
#D(t)= D(t-1)*2/3 + K(t)*1/3  , t = 10, 11, ..., T
```

```python
i=9
while i < len(dataset):
    dataset[i].append(0)
    dataset[i].append(0)
    dataset[i].append(0)
    i+=1
```

```python
i=10
while i < len(dataset):
    if dataset[i][8]>dataset[i][9] and dataset[i][8]>dataset[i-1][8] and dataset[i-1][11]!=1:
        dataset[i][10]=1
    if dataset[i][8]<dataset[i][9] and dataset[i][8]<dataset[i-1][8] and dataset[i-1][11]==1:
        dataset[i][12]=1
    if dataset[i][10]==1 or dataset[i-1][11]==1:
        dataset[i][11]=1
    if dataset[i][12]==1:
        dataset[i][11]=0
        
    i+=1
#KD黃金交叉-建議買進:當KD指標的K值由下往上突破D值，建議買進、做多    K(t) > D(t)
#KD死亡交叉-建議賣出:當KD指標的K值由上往下跌破D值時，建議賣出、做空  K(t) < D(t)
```
